```python
import json
import requests
import pandas as pd
import os
import base64
```

```python
githubToken = os.getenv('GITHUB_TOKEN')

print("Loaded token {}".format(githubToken[0:3]))
```

```python
def getAuth(url, headers={"Authorization":"token {}".format(githubToken)}):
    baseUrl = "https://api.github.com{}".format(url)
    print("Asking {}".format(baseUrl))
    res = requests.get(baseUrl, headers=headers)
    return res.json()
```

```python
data = getAuth('/search/code?q=extension:.scavengerhunt repo:ironhack-datalabs/scavenger')
dicts = data.get('items')
dicts
```

```python
# Append the values of 'name' and 'path' to a new list and sort by file name.

path = []

for dict in dicts:
    path.append(dict['name'] + ';' + dict['path'])
    
path.sort()
path

# Split the sorted strings on ';' to only keep the path.

def split_1(x):
    y = x.split(';')
    return y[1]

path2 = []

for i in path:
    x = split_1(i)
    path2.append(x)
    
print(path2)
```

```python
# Get the contents of each file.

content_summary = []

for item in path2:
    content_sum = getAuth('/repos/ironhack-datalabs/scavenger/contents/{}'.format(item))
    content_summary.append(content_sum)
                    
content_summary
```

```python
# Get the value of 'content' key

content = []

for item in content_summary:
    content.append(item['content'])
    
list(content)
```

```python
# Decode the content codes and convert to strings

content_1 = []

for i in content:
    x = base64.b64decode(i)
    y = str(x)
    content_1.append(y)
    
print(content_1)
```

```python
# Remove charachter 'b' from the left side

content_2 = []

for i in content_1:
    x = i.lstrip('b')    
    content_2.append(x)
print(content_2)
```

```python
# Split each string on '\\n'

content_3 = []

for i in content_2:
    x = i.split("\\n")
    content_3.append(x)

print(content_3)
```

```python
# Remove "'" charachter from each sub-list's first string

content_4 = []

for i in content_3:
    for x in i:
        y = x.lstrip("'")
        content_4.append(y)

print(content_4)
```

```python
# Remove empty strings
clean_text = [x for x in content_4 if x]

# Join text in one string
clean_text_final = ' '.join(clean_text)
print(clean_text_final)
```
